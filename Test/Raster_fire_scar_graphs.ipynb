# Raster fire scar plots

```python
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
pylab.rcParams['figure.figsize'] = (30, 8.0)
```

```python
# enter the file path for the fire scar csv.
df = pd.read_csv(r"Z:\Scratch\Zonal_Stats_Pipeline\Firescars\outputs2\all_Touched_False\birrimba_fire_scar.csv")
```

```python
list(df)
```

```python
reshape_df = df[['year',
 'image',
 'property',
 'laiskey',
 'jan_ha',
 'feb_ha',
 'march_ha',
 'april_ha',
 'may_ha',
 'june_ha',
 'july_ha',
 'aug_ha',
 'sep_ha',
 'oct_ha',
 'nov_ha',
 'dec_ha']]

reshape_df.columns = ['year',
 'image',
 'property',
 'laiskey',
 'January',
 'Febuary',
 'March',
 'April',
 'May',
 'June',
 'July',
 'August',
 'September',
 'October',
 'November',
 'December']
```

```python
reshape_df = reshape_df.set_index('year')
```

```python
reshape_df.plot(kind='bar', stacked=True)
plt.legend(loc='upper right')
plt.title("Property scale fire scars", fontsize=25)
plt.xlabel("Year", fontsize=20)
plt.xticks(rotation=45)
plt.ylabel("Burnt area (ha)", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# to save images out please change path ' T:/rob/' and remove the # below.

#plt.savefig(r'T:/rob/' + 'seasonal_property_scale_fire_scar.png')

#plt.grid(True)

#plt.grid(True)
plt.show()
```

Seasons
In Australia, the seasons are defined by grouping the calendar months in the following way:

Spring - the three transition months September, October and November.
Summer - the three hottest months December, January and February.
Autumn - the transition months March, April and May.
Winter - the three coldest months June, July and August.

```python
reshape_df['Spring']=reshape_df['September']+reshape_df['October']+reshape_df['November']
reshape_df['Summer']=reshape_df['December']+reshape_df['January']+reshape_df['Febuary']
reshape_df['Autumn']=reshape_df['March']+reshape_df['April']+reshape_df['May']
reshape_df['Winter']=reshape_df['June']+reshape_df['July']+reshape_df['August']

# reshape the final dataframe
seasons_df = reshape_df[['Spring', 'Summer', 'Autumn', 'Winter']]
```

```python

seasons_df.plot(kind='bar', stacked=True)
plt.legend(loc='best')
plt.title("Seasonal property scale fire scars", fontsize=25)
plt.xlabel("Year", fontsize=20)
plt.xticks(rotation=45)
plt.ylabel("Burnt area (ha)", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# to save images out please change path ' T:/rob/' and remove the # below.

#plt.savefig(r'T:/rob/' + 'seasonal_property_scale_fire_scar.png')

#plt.grid(True)
plt.show()
```
