```python
# cmdargs
tileA = '106_069'
endfileA = 'dilm2_zstdmask.img'
tileB = '106_070'
endfileB = 'dilm2_zstdmask.img'
```

```python
# Create a list of images
dirname = r'Z:\Landsat\wrs2' + '\\' + tileA
endfilename = endfileA

listA = []
listA_date = []



for root, dirs, files in os.walk(dirname):
    for file in files:
        if file.endswith(endfilename):
            #print(file)
            dateA = file[-27:-19]#+str(2)
            imgA = (os.path.join(root, file))
            listA.append(imgA)
            listA_date.append(dateA)
            

print("list A amount: ", len(listA))
print("Date A amount: ", len(listA_date))
print(listA)
```

```python
# Create a list of images
dirname = r'Z:\Landsat\wrs2' + '\\' + tileB
endfilename = endfileB

listB = []
listB_date = []



for root, dirs, files in os.walk(dirname):
    for file in files:
        if file.endswith(endfilename):
            #print(file)
            dateB = file[-27:-19]#+str(2)
            imgB = (os.path.join(root, file))
            listB.append(imgB)
            listB_date.append(dateB)
            

print("list B amount: ", len(listB))
print("Date B amount: ", len(listB_date))
print(listB)
```
