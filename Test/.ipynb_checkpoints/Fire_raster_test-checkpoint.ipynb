```python
%pylab inline
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import numpy as np
pylab.rcParams['figure.figsize'] = (30, 8.0)
```

```python
df = pd.read_csv(r"Z:\Scratch\Zonal_Stats_Pipeline\Firescars\outputs2\all_Touched_True\birrimba_fire_scar.csv")
```

```python
column_list = ['jan', 'feb', 'march', 'april', 'may', 'june', 'july', 'aug', 'sep', 'oct', 'nov', 'dec']
for col in column_list:
    month_area = str(col)+'_area_ha'
    col_ =  str(col)
    df[month_area] = (df[col_]*(250*250)/10000).astype(int) 
```

```python
# Fill nall values with 0
df.fillna(0, inplace=True)
# Add columns filled with 0 to standardise each df
column_list =['unburnt', 'jan', 'feb', 'march', 'april', 'may', 'june', 'july', 'aug', 'sep', 'oct', 'nov', 'dec', '255.0']

for col in column_list:
    if col not in df.columns:
        df[col] = 0
# Reorder the dataframe        
clean_fire = df[['year', 'image', 'property','uid', 'laiskey', 'count', 'unburnt', 'jan', 'feb', 'march', 'april', 'may', 'june', 'july', 'aug', 'sep', 'oct', 'nov', 'dec', '255.0', 'WGSz52_ha', 'WGSz53_ha']]

# convert count values from float to int
clean_fire[['count', 'unburnt', 'jan', 'feb', 'march', 'april', 'may', 'june', 'july', 'aug', 'sep', 'oct', 'nov', 'dec', '255.0']] = clean_fire[['count', 'unburnt', 'jan', 'feb', 'march', 'april', 'may', 'june', 'july', 'aug', 'sep', 'oct', 'nov', 'dec', '255.0']].astype(int)


# NAFI null values (255) are 0 values in other years.
clean_fire['unburnt'] = clean_fire['unburnt']+clean_fire['255.0']
clean_fire.drop(['255.0'], inplace = True, axis=1)
# Calculate the area based on the count values
clean_fire["count_area_ha"] = (clean_fire["count"]*(250*250)/10000)
```
